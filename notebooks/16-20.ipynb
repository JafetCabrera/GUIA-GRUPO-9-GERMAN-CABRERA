# Ejercicio 16: DataFrame de estudiantes 

```python
import pandas as pd

data = {
    'Fecha': ['2026-06-01', '2026-06-01', '2026-06-02', '2026-06-02', '2026-06-03',
              '2026-06-03', '2026-06-04', '2026-06-04', '2026-06-05', '2026-06-05',
              '2026-06-06', '2026-06-06', '2026-06-07', '2026-06-07', '2026-06-07'],
    'Vendedor': ['Diego', 'Laura', 'Andrés', 'Diego', 'Laura', 'Andrés', 'Diego', 'Laura', 
                 'Andrés', 'Diego', 'Laura', 'Andrés', 'Diego', 'Laura', 'Andrés'],
    'Producto': ['Taladro', 'Martillo', 'Caja de Clavos', 'Destornillador', 'Sierra Circular', 
                 'Cinta Aislante', 'Martillo', 'Caja de Clavos', 'Taladro', 'Cinta Aislante', 
                 'Destornillador', 'Sierra Circular', 'Caja de Clavos', 'Taladro', 'Martillo'],
    'Categoría': ['Eléctricas', 'Manuales', 'Consumibles', 'Manuales', 'Eléctricas', 
                  'Consumibles', 'Manuales', 'Consumibles', 'Eléctricas', 'Consumibles', 
                  'Manuales', 'Eléctricas', 'Consumibles', 'Eléctricas', 'Manuales'],
    'Cantidad': [2, 5, 10, 8, 1, 15, 3, 12, 1, 20, 6, 2, 8, 3, 4],
    'Precio Unitario': [150, 15, 5, 8, 120, 2, 15, 5, 150, 2, 8, 120, 5, 150, 15]
}

df_ventas = pd.DataFrame(data)

df_ventas['TotalVenta'] = df_ventas['Cantidad'] * df_ventas['Precio Unitario']

print("--- DataFrame con Nuevos Datos (Ferretería) ---")
print(df_ventas.head(15))
print("\n" + "-"*60 + "\n")

print("--- Total vendido por vendedor ---")
total_por_vendedor = df_ventas.groupby('Vendedor')['TotalVenta'].sum()
print(total_por_vendedor)
print("\n" + "-"*60 + "\n")

print("--- Producto más vendido (por unidades) ---")
producto_mas_vendido = df_ventas.groupby('Producto')['Cantidad'].sum().idxmax()
unidades_producto = df_ventas.groupby('Producto')['Cantidad'].sum().max()
print(f"El producto más vendido es '{producto_mas_vendido}' con {unidades_producto} unidades.")
print("\n" + "-"*60 + "\n")

print("--- Categoría con mayor ingreso ---")
categoria_mayor_ingreso = df_ventas.groupby('Categoría')['TotalVenta'].sum().idxmax()
ingreso_categoria = df_ventas.groupby('Categoría')['TotalVenta'].sum().max()
print(f"La categoría con mayor ingreso es '{categoria_mayor_ingreso}' con un total de L. {ingreso_categoria:.2f}")
print("\n" + "-"*60 + "\n")

print("--- Ventas ordenadas de mayor a menor ---")
df_ordenado = df_ventas.sort_values(by='TotalVenta', ascending=False)
print(df_ordenado)
```

# Ejercicio 18: Limpieza de datos con pandas 

```python
import pandas as pd

data = {
    'Nombre': ['Laura', 'Andrés', 'Carmen', 'David', 'Sofía', 'Miguel', 'Lucía', 'Diego'],
    'Edad': [25, None, 30, 22, None, 45, 28, 35],
    'Correo': ['laura@email.com', 'andres@email.com', None, 'david@email.com', 'sofia@email.com', None, 'lucia@email.com', 'diego@email.com'],
    'Ciudad': ['San Pedro Sula', 'Tegucigalpa', None, 'La Ceiba', None, 'Comayagua', 'Choloma', 'El Progreso'],
    'Teléfono': ['99887766', None, '88776655', '77665544', '66554433', None, '55443322', '44332211']
}

df_clientes = pd.DataFrame(data)

print("--- DataFrame Original (Con datos incompletos) ---")
print(df_clientes)
print("\n" + "-"*60 + "\n")

print("--- Identificación de valores nulos (True/False) ---")
print(df_clientes.isnull())
print("\n" + "-"*60 + "\n")

print("--- Conteo de valores nulos por columna ---")
print(df_clientes.isnull().sum())
print("\n" + "-"*60 + "\n")

df_clientes['Ciudad'] = df_clientes['Ciudad'].fillna('No especificada')

df_clientes = df_clientes.dropna(subset=['Correo'])

print("--- DataFrame Limpio ---")
print(df_clientes)
```

# Ejercicio 19: Análisis de préstamos

```python
import pandas as pd

data = {
    'Cliente': ['Juan', 'María', 'Pedro', 'Ana', 'Luis', 'Carmen', 'Jorge', 'Rosa', 'Carlos', 'Lucía'],
    'Monto Solicitado': [45000, 60000, 20000, 75000, 30000, 55000, 15000, 80000, 40000, 90000],
    'Plazo': [12, 24, 6, 36, 12, 24, 6, 48, 12, 60],
    'Ingreso Mensual': [5000, 8000, 3000, 10000, 4000, 7000, 2500, 12000, 4500, 15000],
    'Estado de Solicitud': ['Aprobado', 'Pendiente', 'Aprobado', 'Rechazado', 'Aprobado', 'Pendiente', 'Aprobado', 'Rechazado', 'Pendiente', 'Aprobado']
}

df_prestamos = pd.DataFrame(data)

print("--- DataFrame de Préstamos ---")
print(df_prestamos)
print("\n" + "-"*60 + "\n")

print("--- Monto Promedio Solicitado ---")
promedio_monto = df_prestamos['Monto Solicitado'].mean()
print(f"L. {promedio_monto:.2f}")
print("\n" + "-"*60 + "\n")

print("--- Conteo de Solicitudes por Estado ---")
conteo_estados = df_prestamos['Estado de Solicitud'].value_counts()
print(conteo_estados)
print("\n" + "-"*60 + "\n")

print("--- Préstamos Aprobados ---")
aprobados = df_prestamos[df_prestamos['Estado de Solicitud'] == 'Aprobado']
print(aprobados)
print("\n" + "-"*60 + "\n")

print("--- Clientes con monto solicitado mayor a 50,000 ---")
mayor_50k = df_prestamos[df_prestamos['Monto Solicitado'] > 50000]
print(mayor_50k)
print("\n" + "-"*60 + "\n")

print("--- Promedio de Monto Solicitado por Estado ---")
promedio_por_estado = df_prestamos.groupby('Estado de Solicitud')['Monto Solicitado'].mean()
print(promedio_por_estado)
```

# Ejercicio 20: Mini análisis de datos tipo empresa 

```python
import pandas as pd

data = {
    'Nombre': ['Ana', 'Carlos', 'Luis', 'Marta', 'Jorge', 'Elena', 'Pedro', 'Sofia', 'Raul', 'Carmen'],
    'Departamento': ['TI', 'Ventas', 'Seguridad', 'TI', 'Finanzas', 'Proyectos', 'TI', 'RRHH', 'Dirección', 'Ventas'],
    'Cargo': ['Desarrollador Web', 'Asesor Comercial', 'Auditor IT', 'Administrador Linux', 'Analista', 'Gerente de Proyectos', 'Soporte Técnico', 'Reclutadora', 'Director General', 'Asesor Comercial'],
    'Salario': [25000, 15000, 32000, 28000, 22000, 40000, 14000, 16000, 65000, 15500],
    'Años Experiencia': [3, 1, 5, 6, 4, 8, 2, 3, 15, 2],
    'Ciudad': ['San Pedro Sula', 'Tegucigalpa', 'La Ceiba', 'San Pedro Sula', 'Choluteca', 'Tegucigalpa', 'Comayagua', 'San Pedro Sula', 'Tegucigalpa', 'La Ceiba']
}

df_empleados = pd.DataFrame(data)

print("--- Primeros registros ---")
print(df_empleados.head())
print("\n" + "-"*60 + "\n")

print("--- Estadísticas generales de salarios ---")
print(df_empleados['Salario'].describe())
print("\n" + "-"*60 + "\n")

print("--- Salario promedio por departamento ---")
promedio_dept = df_empleados.groupby('Departamento')['Salario'].mean()
print(promedio_dept)
print("\n" + "-"*60 + "\n")

print("--- Empleado con salario más alto ---")
empleado_top = df_empleados.loc[df_empleados['Salario'].idxmax()]
print(empleado_top)
print("\n" + "-"*60 + "\n")

print("--- Empleados con más de 5 años de experiencia ---")
mas_5_anios = df_empleados[df_empleados['Años Experiencia'] > 5]
print(mas_5_anios)
print("\n" + "-"*60 + "\n")

print("--- Empleados ordenados por salario (Mayor a Menor) ---")
df_ordenado = df_empleados.sort_values(by='Salario', ascending=False)
print(df_ordenado)
print("\n" + "-"*60 + "\n")


def asignar_nivel(anios):
    if anios <= 3:
        return 'Junior'
    elif anios <= 7:
        return 'Intermedio'
    else:
        return 'Senior'

df_empleados['NivelExperiencia'] = df_empleados['Años Experiencia'].apply(asignar_nivel)

print("--- DataFrame final con NivelExperiencia ---")
print(df_empleados)
```
