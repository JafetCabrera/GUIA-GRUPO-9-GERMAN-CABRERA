```python
# Lista de productos
productos = [
    {"codigo": 1, "nombre": "Laptop", "precio": 15000, "cantidad": 2},
    {"codigo": 2, "nombre": "Mouse", "precio": 500, "cantidad": 5},
    {"codigo": 3, "nombre": "Teclado", "precio": 1200, "cantidad": 3}
]

# Mostrar productos
for producto in productos:
    print(producto)

# Buscar producto por código
codigo = 2

for producto in productos:
    if producto["codigo"] == codigo:
        print("\nProducto encontrado:")
        print(producto)

# Calcular total del inventario
total = 0

for producto in productos:
    total += producto["precio"] * producto["cantidad"]

print("\nValor total del inventario:", total)
```

Explicación
Se creó una lista de productos usando diccionarios.  
Después se mostraron los productos, se buscó uno por código y se calculó el total del inventario.
